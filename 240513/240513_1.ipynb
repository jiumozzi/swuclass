```python
data = dict()

with open("NC_045512.fasta") as handle:
    for line in handle:
        if line.startswith(">"):
            continue

        for base in line.strip():
            if base not in data:
                data[base] = 0

            data[base] += 1

print(data)
```

```python
import pandas as pd

df_data = {"base": data.keys(), "base_count": data.values()}
print(df_data)

df = pd.DataFrame(df_data)
df.plot(
    kind="bar",
    x="base",
    y="base_count",
    title="Base count",
    xlabel="Base",
    ylabel="Count",
)
```
